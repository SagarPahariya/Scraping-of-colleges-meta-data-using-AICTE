```python
import pandas as pd
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.maximize_window()
driver.get("https://facilities.aicte-india.org/dashboard/pages/angulardashboard.php#!/approved")
time.sleep(2)
```

```python
a=pd.read_excel('states.xlsx')
```

```python
col1=a['states'].values.tolist()
```

```python
c_name,c_add ,c_distict =[],[],[]
```

```python
add_list=[] # original list of address
```

```python
for s in col1:
    select = Select(driver.find_element(By.XPATH,'//*[@id="state"]'))
    select.select_by_visible_text(s)
    time.sleep(2)
    driver.find_element(By.XPATH,'//*[@id="load"]').click()
    time.sleep(2)
    s0=driver.page_source
    src1=BeautifulSoup(s0,'html')
    name11=src1.find('div', {'class' : 'w3-main'})
    n22=name11.find('div',{'class':'ng-scope'})
    n99=n22.find('div',{'class':'w3-main w3-container ng-scope'})
    n100=n99.find('div',{'id':'jsontable_wrapper'})
    n010=n100.find('span')
    n020=n010.find_all('a')
    n030=len(n020)-1
    n040=n020[n030]
    n050=n040.text
    n050=int(n050)    
    for m in range(n050):
        s=driver.page_source
        src=BeautifulSoup(s,'html')
        name1=src.find('div', {'class' : 'w3-main'})
        n2=name1.find('div',{'class':'ng-scope'})
        n9=n2.find('div',{'class':'w3-main w3-container ng-scope'})
        n10=n9.find('div',{'id':'jsontable_wrapper'})
        n11=n10.find('table',{'id':'jsontable'})
        n12=n11.find('tbody') # body
        n13=n12.find_all('tr')
        for i in range(len(n13)):
            n14=n13[i]
            n15=n14.find_all('td')
            c_name.append(n15[1].text)
            c_add.append(n15[2].text)
            c_distict.append(n15[3].text)
        driver.find_element(By.XPATH,'//*[@id="jsontable_next"]').click()
```

```python
for j in range(len(c_add)):
    a=c_add[j].partition('\n')
    b=a[2].replace('\n', ' ')
    mn=a[0]+b #main name
    add_list.append(mn)
```

```python
len(add_list)
```

```python
data={'College Name': c_name , 'College address' : add_list, 'College District' : c_distict}
```

```python
df=pd.DataFrame(data=data)
```
